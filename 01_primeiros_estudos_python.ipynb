# 7. Lista de dicionários

Em sistemas reais, normalmente temos vários registros.

Por exemplo: vários alunos, cada um com seus dados.

```python
alunos = [
    {
        "nome": "Ana",
        "serie": "9º ano",
        "disciplina": "Matemática",
        "valor_aula": 130.00,
        "aulas_por_semana": 2
    },
    {
        "nome": "João",
        "serie": "8º ano",
        "disciplina": "Ciências",
        "valor_aula": 120.00,
        "aulas_por_semana": 1
    },
    {
        "nome": "Maria",
        "serie": "1ª série EM",
        "disciplina": "Física",
        "valor_aula": 150.00,
        "aulas_por_semana": 1
    }
]

print(alunos)
```

```python
for aluno in alunos:
    print("Nome:", aluno["nome"])
    print("Disciplina:", aluno["disciplina"])
    print("Valor da aula: R$", aluno["valor_aula"])
    print("-" * 30)
```

# 8. Calculando o valor mensal de cada aluno

Agora vou usar os dados de cada aluno para calcular quanto cada um paga por mês.

A fórmula usada será:

valor mensal = valor da aula × aulas por semana × 4

```python
for aluno in alunos:
    valor_mensal = aluno["valor_aula"] * aluno["aulas_por_semana"] * 4
    
    print("Nome:", aluno["nome"])
    print("Valor mensal: R$", valor_mensal)
    print("-" * 30)
```

# 9. Criando uma função para calcular o valor mensal

Funções servem para organizar o código e evitar repetição.

Agora vou criar uma função que recebe:

- valor da aula;
- quantidade de aulas por semana;

e devolve o valor mensal estimado.

```python
def calcular_valor_mensal(valor_aula, aulas_por_semana):
    valor_mensal = valor_aula * aulas_por_semana * 4
    return valor_mensal
```

```python
resultado = calcular_valor_mensal(130.00, 2)

print("Valor mensal: R$", resultado)
```

```python
for aluno in alunos:
    valor_mensal = calcular_valor_mensal(
        aluno["valor_aula"],
        aluno["aulas_por_semana"]
    )
    
    print("Nome:", aluno["nome"])
    print("Valor mensal: R$", valor_mensal)
    print("-" * 30)
```

# 10. Gerando um relatório completo dos alunos

Agora vou criar uma função que percorre a lista de alunos, calcula o valor mensal de cada um e mostra um relatório organizado.

Essa função será o primeiro passo para transformar o código em um sistema de controle de aulas.

```python
def gerar_relatorio_alunos(lista_alunos):
    print("RELATÓRIO DE ALUNOS")
    print("=" * 40)
    
    total_mensal = 0
    
    for aluno in lista_alunos:
        valor_mensal = calcular_valor_mensal(
            aluno["valor_aula"],
            aluno["aulas_por_semana"]
        )
        
        total_mensal = total_mensal + valor_mensal
        
        print("Nome:", aluno["nome"])
        print("Série:", aluno["serie"])
        print("Disciplina:", aluno["disciplina"])
        print("Valor por aula: R$", aluno["valor_aula"])
        print("Aulas por semana:", aluno["aulas_por_semana"])
        print("Valor mensal: R$", valor_mensal)
        print("-" * 40)
    
    print("Total mensal estimado: R$", total_mensal)
        
```

```python
gerar_relatorio_alunos(alunos)
```
