# 2. Função para calcular valor mensal

Esta função recebe o valor de uma aula e a quantidade de aulas por semana.

Depois, calcula o valor mensal estimado considerando 4 semanas.

```python
def calcular_valor_mensal(valor_aula, aulas_por_semana):
    valor_mensal = valor_aula * aulas_por_semana * 4
    return valor_mensal
```

```python
teste = calcular_valor_mensal(130.00, 2)

print("Valor mensal de teste: R$", teste)
```

# 3. Função para cadastrar aluno

Agora vou criar uma função que recebe os dados de um aluno e adiciona esse aluno à lista.

```python
alunos = [
    {
        "nome": "Ana",
        "serie": "9º ano",
        "disciplina": "Matemática",
        "valor_aula": 130.00,
        "aulas_por_semana": 2
    },
    {
        "nome": "João",
        "serie": "8º ano",
        "disciplina": "Ciências",
        "valor_aula": 120.00,
        "aulas_por_semana": 1
    },
    {
        "nome": "Maria",
        "serie": "1ª série EM",
        "disciplina": "Física",
        "valor_aula": 150.00,
        "aulas_por_semana": 1
    }
]
```

# 4. Salvando dados em arquivo JSON

Agora vou salvar a lista de alunos em um arquivo `.json`.

Isso permite guardar os dados mesmo depois de fechar o Jupyter.

```python
def salvar_alunos_json(lista_alunos, nome_arquivo):
    with open(nome_arquivo,"w", encoding="utf-8") as arquivo:
        json.dump(lista_alunos, arquivo, ensure_ascii=False, indent=4)

    print("dados salvos com sucesso", nome_arquivo)
    
```

```python
salvar_alunos_json(alunos, "alunos.json")
```

# 5. Carregando dados de arquivo JSON

Agora vou criar uma função para carregar os dados salvos no arquivo `alunos.json`.

Isso permite recuperar os alunos mesmo depois de fechar o Jupyter.

```python
def carregar_alunos_json(nome_arquivo):
    with open(nome_arquivo, "r", encoding="utf-8") as arquivo:
        dados = json.load(arquivo)
    
    print("Dados carregados com sucesso de:", nome_arquivo)
    return dados
```

```python
alunos_carregados = carregar_alunos_json("alunos.json")

print(alunos_carregados)
```
