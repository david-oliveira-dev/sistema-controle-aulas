# Notebook 03 — Sistema com Menu

Neste notebook, vou começar a transformar o sistema de controle de aulas em um programa mais organizado.

A ideia é criar um menu com opções para cadastrar alunos, listar alunos, gerar relatório, salvar dados e carregar dados.

## Objetivos

- criar um menu de opções;
- usar `input()` para entrada de dados;
- cadastrar alunos de forma interativa;
- listar alunos cadastrados;
- gerar relatório financeiro;

- aproximar o código de um sistema real.

```python
import json
```

# 1. Funções do sistema

Agora vou reutilizar as funções criadas anteriormente para montar um sistema completo.

```python
def caucular_valor_mensal(valor_aula, aulas_por_semana):
    valor_mensal = valor_aula*aulas_por_semana* 4 
    return valor_mensal 
```

```python
def cadastrar_aluno(lista_alunos, nome, serie, disciplina, valor_aula, aulas_por_semana):
    novo_aluno = {
        "nome": nome,
        "serie": serie,
        "disciplina": disciplina,
        "valor_aula": valor_aula,
        "aulas_por_semana": aulas_por_semana
    }
    
    lista_alunos.append(novo_aluno)
    
    print("Aluno cadastrado com sucesso!")
    
```

```python
def gerar_relatorio_alunos(lista_alunos):
    print("\nRELATORIO DE ALUNOS")
    print("=" * 40 )

    total_mensal = 0 

    for aluno in lista_alunos:
        valor_mensal = caucular_valor_mensal(
            aluno["valor_aula"],
            aluno["aulas_por_semana"]
        )

        total_mensal+= valor_mensal 

        print(f"Nome: {aluno['nome']}")
        print(f"Série: {aluno['serie']}")
        print(f"Disciplina: {aluno['disciplina']}")
        print(f"Valor por aula: R$ {aluno['valor_aula']:.2f}")
        print(f"Aulas por semana: {aluno['aulas_por_semana']}")
        print(f"Valor mensal: R$ {valor_mensal:.2f}")
        print("-" * 40)
    
    print(f"Total mensal estimado: R$ {total_mensal:.2f}")
        
```

```python
def salvar_alunos_json(lista_alunos, nome_arquivo):
    with open(nome_arquivo,"w", encoding="utf-8") as arquivo:
        json.dump(lista_alunos, arquivo, ensure_ascii=False, indent=4)

    print("Dados Salvos com sucesso")
```

```python
def carregar_alunos_json(nome_arquivo):
    try:
        with open(nome_arquivo, "r", encoding="utf-8") as arquivo:
            dados = json.load(arquivo) 

        print("Dados carregados com sucesso")
        return dados

    except FileNotFoundError:
        print("Arquivo nao encontrado.")
        return[]
        
```

# 2. Lista inicial de alunos

Agora vou carregar os alunos do arquivo JSON.

```python
alunos = carregar_alunos_json("alunos.json")
```

# 3. Criando o menu do sistema

Agora vou criar um menu simples usando `while`.

```python
while True:
    print("\nSISTEMA DE CONTROLE DE AULAS")
    print("=" * 40)
    print("1 - Cadastrar aluno")
    print("2 - Listar alunos")
    print("3 - Gerar relatório")
    print("4 - Salvar dados")
    print("5 - Remover aluno")
    print("6 - Editar aluno")
    print("0 - Sair")

    opcao = input("Escolha uma opção ")

    if opcao == "0":
        print("Encerrando sistema")
        break
```

# 4. Cadastro interativo de alunos

Agora o sistema vai permitir cadastrar alunos usando `input()`.

```python
while True:
    print("\nSISTEMA DE CONTROLE DE AULAS")
    print("=" * 40)
    print("1 - Cadastrar aluno")
    print("2 - Listar alunos")
    print("3 - Gerar relatório")
    print("4 - Salvar dados")
    print("5 - Remover aluno")
    print("6 - Editar aluno")
    print("0 - Sair")
    opcao = input("Escolha uma opção: ")
    
    if opcao == "1":
        print("\nCADASTRO DE ALUNO")
        print("=" * 40)
        
        nome = input("Nome do aluno: ")
        serie = input("Série: ")
        disciplina = input("Disciplina: ")
        
        valor_aula = float(input("Valor da aula: "))
        
        aulas_por_semana = int(
            input("Aulas por semana: ")
        )
        
        cadastrar_aluno(
            alunos,
            nome,
            serie,
            disciplina,
            valor_aula,
            aulas_por_semana
        )
    
    elif opcao == "0":
        print("Encerrando sistema...")
        break
```

# 6. Funcionalidade de geração de relatório

Agora o sistema será capaz de gerar um relatório completo dos alunos cadastrados diretamente pelo menu.

Nesta etapa, vou reutilizar a função `gerar_relatorio_alunos()` criada anteriormente.

O relatório exibirá:
- nome do aluno;
- série;
- disciplina;
- valor da aula;
- quantidade de aulas por semana;
- valor mensal estimado;
- total mensal do sistema.

Essa funcionalidade aproxima o projeto de um sistema administrativo real.

```python
while True:
    print("\nSISTEMA DE CONTROLE DE AULAS")
    print("=" * 40)
    print("1 - Cadastrar aluno")
    print("2 - Listar alunos")
    print("3 - Gerar relatório")
    print("4 - Salvar dados")
    print("5 - Remover aluno")
    print("6 - Editar aluno")
    print("0 - Sair")
    
    opcao = input("Escolha uma opção: ")
    
    if opcao == "1":
        print("\nCADASTRO DE ALUNO")
        print("=" * 40)
        
        nome = input("Nome do aluno: ")
        serie = input("Série: ")
        disciplina = input("Disciplina: ")
        
        valor_aula = float(
            input("Valor da aula: ")
        )
        
        aulas_por_semana = int(
            input("Aulas por semana: ")
        )
        
        cadastrar_aluno(
            alunos,
            nome,
            serie,
            disciplina,
            valor_aula,
            aulas_por_semana
        )
    
    elif opcao == "2":
        print("\nLISTA DE ALUNOS")
        print("=" * 40)
        
        if len(alunos) == 0:
            print("Nenhum aluno cadastrado.")
        
        else:
            for aluno in alunos:
                print(f"Nome: {aluno['nome']}")
                print(f"Série: {aluno['serie']}")
                print(f"Disciplina: {aluno['disciplina']}")
                print("-" * 40)
    
    elif opcao == "3":
        print("\nGERANDO RELATÓRIO...")
        
        if len(alunos) == 0:
            print("Nenhum aluno cadastrado.")
        
        else:
            gerar_relatorio_alunos(alunos)
    
    elif opcao == "0":
        print("Encerrando sistema...")
        break
```

# 7. Funcionalidade de salvamento de dados

Agora o sistema será capaz de salvar os dados diretamente pelo menu.

Nesta etapa, vou reutilizar a função `salvar_alunos_json()` criada anteriormente.

Essa funcionalidade permite:
- armazenar os dados dos alunos;
- manter os dados após fechar o programa;
- atualizar o arquivo JSON sempre que necessário.

Esse é um passo importante para transformar o projeto em um sistema funcional.

```python
while True:
    print("\nSISTEMA DE CONTROLE DE AULAS")
    print("=" * 40)
    print("1 - Cadastrar aluno")
    print("2 - Listar alunos")
    print("3 - Gerar relatório")
    print("4 - Salvar dados")
    print("5 - Remover aluno")
    print("6 - Editar aluno")
    print("0 - Sair")
    
    opcao = input("Escolha uma opção: ")
    
    if opcao == "1":
        print("\nCADASTRO DE ALUNO")
        print("=" * 40)
        
        nome = input("Nome do aluno: ")
        serie = input("Série: ")
        disciplina = input("Disciplina: ")
        
        valor_aula = float(
            input("Valor da aula: ")
        )
        
        aulas_por_semana = int(
            input("Aulas por semana: ")
        )
        
        cadastrar_aluno(
            alunos,
            nome,
            serie,
            disciplina,
            valor_aula,
            aulas_por_semana
        )
    
    elif opcao == "2":
        print("\nLISTA DE ALUNOS")
        print("=" * 40)
        
        if len(alunos) == 0:
            print("Nenhum aluno cadastrado.")
        
        else:
            for aluno in alunos:
                print(f"Nome: {aluno['nome']}")
                print(f"Série: {aluno['serie']}")
                print(f"Disciplina: {aluno['disciplina']}")
                print("-" * 40)
    
    elif opcao == "3":
        print("\nGERANDO RELATÓRIO...")
        
        if len(alunos) == 0:
            print("Nenhum aluno cadastrado.")
        
        else:
            gerar_relatorio_alunos(alunos)
    
    elif opcao == "4":
        salvar_alunos_json(alunos, "alunos.json")
    
    elif opcao == "0":
        print("Encerrando sistema...")
        break
```

# 8. Funcionalidades de remoção e edição de alunos

Agora o sistema será capaz de:
- remover alunos cadastrados;
- editar informações de alunos existentes.

Essas funcionalidades aproximam o projeto de um CRUD completo.

Nesta etapa, serão utilizados:
- busca em listas;
- alteração de dicionários;
- remoção de elementos;
- estruturas condicionais;
- atualização dinâmica de dados.

```python
while True:

    print("\nSISTEMA DE CONTROLE DE AULAS")
    print("=" * 40)
    print("1 - Cadastrar aluno")
    print("2 - Listar alunos")
    print("3 - Gerar relatório")
    print("4 - Salvar dados")
    print("5 - Remover aluno")
    print("0 - Sair")

    opcao = input("Escolha uma opção: ")

    if opcao == "1":

        print("\nCADASTRO DE ALUNO")
        print("=" * 40)

        nome = input("Nome do aluno: ")
        serie = input("Série: ")
        disciplina = input("Disciplina: ")

        valor_aula = float(
            input("Valor da aula: ")
        )

        aulas_por_semana = int(
            input("Aulas por semana: ")
        )

        cadastrar_aluno(
            alunos,
            nome,
            serie,
            disciplina,
            valor_aula,
            aulas_por_semana
        )



    elif opcao == "2":

        print("\nLISTA DE ALUNOS")
        print("=" * 40)

        if len(alunos) == 0:

            print("Nenhum aluno cadastrado.")

        else:

            for aluno in alunos:

                print(f"Nome: {aluno['nome']}")
                print(f"Série: {aluno['serie']}")
                print(f"Disciplina: {aluno['disciplina']}")

                print("-" * 40)

    elif opcao == "3":

        print("\nGERANDO RELATÓRIO")

        if len(alunos) == 0:

            print("Nenhum aluno cadastrado.")

        else:

            gerar_relatorio_alunos(alunos)


    elif opcao == "4":

        salvar_alunos_json(
            alunos,
            "alunos.json"
        )


    elif opcao == "5":

        print("\nREMOVER ALUNO")
        print("=" * 40)

        nome_remover = input(
            "Digite o nome do aluno: "
        )

        aluno_encontrado = False

        for aluno in alunos:

            if aluno["nome"] == nome_remover:

                alunos.remove(aluno)

                print(
                    "Aluno removido com sucesso!"
                )

                aluno_encontrado = True

                break

        if aluno_encontrado == False:

            print("Aluno não encontrado.")

 

    elif opcao == "0":

        print("Encerrando sistema...")

        break

    else:

        print("Opção inválida.")
```

# 9. Funcionalidade de edição de alunos

Agora o sistema será capaz de editar informações de alunos já cadastrados.

Nesta etapa, serão utilizados:
- busca em listas;
- alteração de dicionários;
- estruturas condicionais;
- atualização dinâmica de dados.

Essa funcionalidade aproxima o projeto de um CRUD completo.

```python
while True:

    print("\nSISTEMA DE CONTROLE DE AULAS")
    print("=" * 40)
    print("1 - Cadastrar aluno")
    print("2 - Listar alunos")
    print("3 - Gerar relatório")
    print("4 - Salvar dados")
    print("5 - Remover aluno")
    print("6 - Editar aluno")
    print("0 - Sair")

    opcao = input("Escolha uma opção: ")


    if opcao == "1":

        print("\nCADASTRO DE ALUNO")
        print("=" * 40)

        nome = input("Nome do aluno: ")
        serie = input("Série: ")
        disciplina = input("Disciplina: ")

        valor_aula = float(
            input("Valor da aula: ")
        )

        aulas_por_semana = int(
            input("Aulas por semana: ")
        )

        cadastrar_aluno(
            alunos,
            nome,
            serie,
            disciplina,
            valor_aula,
            aulas_por_semana
        )


    elif opcao == "2":

        print("\nLISTA DE ALUNOS")
        print("=" * 40)

        if len(alunos) == 0:

            print("Nenhum aluno cadastrado.")

        else:

            for aluno in alunos:

                print(f"Nome: {aluno['nome']}")
                print(f"Série: {aluno['serie']}")
                print(f"Disciplina: {aluno['disciplina']}")

                print("-" * 40)



    elif opcao == "3":

        print("\nGERANDO RELATÓRIO")

        if len(alunos) == 0:

            print("Nenhum aluno cadastrado.")

        else:

            gerar_relatorio_alunos(alunos)


    elif opcao == "4":

        salvar_alunos_json(
            alunos,
            "alunos.json"
        )



    elif opcao == "5":

        print("\nREMOVER ALUNO")
        print("=" * 40)

        nome_remover = input(
            "Digite o nome do aluno: "
        )

        aluno_encontrado = False

        for aluno in alunos:

            if aluno["nome"] == nome_remover:

                alunos.remove(aluno)

                print(
                    "Aluno removido com sucesso!"
                )

                aluno_encontrado = True

                break

        if aluno_encontrado == False:

            print("Aluno não encontrado.")


    elif opcao == "6":

        print("\nEDITAR ALUNO")
        print("=" * 40)

        nome_busca = input(
            "Digite o nome do aluno: "
        )

        aluno_encontrado = False

        for aluno in alunos:

            if aluno["nome"] == nome_busca:

                print("\nAluno encontrado!")
                print("=" * 40)

                print(f"Nome atual: {aluno['nome']}")
                print(f"Série atual: {aluno['serie']}")
                print(
                    f"Disciplina atual: "
                    f"{aluno['disciplina']}"
                )

                aluno["nome"] = input(
                    "Novo nome: "
                )

                aluno["serie"] = input(
                    "Nova série: "
                )

                aluno["disciplina"] = input(
                    "Nova disciplina: "
                )

                aluno["valor_aula"] = float(
                    input("Novo valor da aula: ")
                )

                aluno["aulas_por_semana"] = int(
                    input("Novas aulas por semana: ")
                )

                print(
                    "\nAluno atualizado com sucesso!"
                )

                aluno_encontrado = True

                break

        if aluno_encontrado == False:

            print("Aluno não encontrado.")


    elif opcao == "0":

        print("Encerrando sistema...")

        break


    else:

        print("Opção inválida.")
```
